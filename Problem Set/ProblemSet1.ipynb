# Problem 1
## Question
Translate the following sentences from English to predicate logic.  
$X$ is the set of people  
$S(x)$ means x is a student of 6.042  
$A(x)$ means x has gotten an 'A' in 6.042  
$T(x)$ means x is a TA of 6.042  
$E(x,y)$ means x and y are the same person
  
**(a)** There are people who have taken 6.042 and have gotten A's in 6.042  
**(b)** All people who are TA's and have taken 6.042 got A's in 6.042  
**(c)** There are no people who are 6.042 TA's who did not get A's in 6.042  
**(d)** There are at least three people who are TA's in 6.042 and have not taken 6.042  
  
*Hint*: $\wedge$ means *And*  ||| $\vee$ means *Or*  ||| $\neg or \bar{X}$ means *Not*

## Answer  
**(a)** The sentence tells that there exists people that belongs to $X$ who belongs to $S(x)$ and $A(x)$.  
*Solution* $$\exists x\in X\mid x\in (S(x)\wedge A(x))$$  
or $$S(x)\wedge A(x)\notin \emptyset$$  
**(b)** The sentence tells that everyone in $T(x)$ and also in $S(x)$ is in $A(x)$  
*Solution* $$T(x)\wedge S(x)\in A(x)$$  
or $$(T(x)\cup S(x))\subset A(x)$$  
**(c)** The sentence tells that nopeople is in $T(x)$ and not in $A(x)$  
*Solution* $$T(x)\wedge (\neg A(x))\in \emptyset$$  
**(d)** The sentence tells that there are three people who is in $T(x)$ and not in $S(x)$ and they are not the same person  
*Solution* $$\exists x,y,z\in X\mid (T(x)\wedge \neg S(x))\wedge (T(y)\wedge \neg S(y))\wedge(T(z)\wedge \neg S(z))\wedge (\neg E(x,y) \wedge \neg E(y,z) \wedge \neg E(z,x))$$

# Problem 4
## Question
Given 12 coins and a balance scale, one of the coins is fake that weigh less than other. Find the fake one within 3 weighings.

## Answer 
Using bi-section search to narrow the scale.  
*solution* (1)The first weighing divided the 12 coins into two 6-coins sets, and after weighing these two sets, we can easily get the set with a fake coin.  
(2)The second step is similar except that we divided the 6-coins set with a fake coin into two 3-coins set and find the set with a fake coin.  
(3)The last step is that we randomly choose two coins from the 3-coins set. If the fake coin being chosen, then we can tell the fake coin by just weighing the two coins. Else, if both coins we choose are real, then we can tell that the balance scale won't tip, so the fake one must be the one not being chosen.

# Problem 5 
## Question
Prove the following statement by proving its contrapositive: if $r$ is irrational, then $r^{1\over 5}$ is irrational.  
  
## Answer
The contrapositive statement is: if $r^{1\over 5}$ is not irrational, then $r$ is not irrational. In other word, if $r^{1\over 5}$ is rational, then $r$ is rational.  
*Solution* Given that $r^{1\over 5}$ is rational, then it can be represented as a fraction say $p\over q$ and p,q are relatively prime.  
If so, we can also represent $r$ as a fraction $m\over n$ as $m = p^{5}, n = q^{5}$  
$$Q.E.D.$$

# Problem 6
## Question
Suppose that $w^{2}+x^{2}+y^{2}=z^{2}$, where w,x,y, and z always denote positive integers.  
Prove the proposition: z is even iff w,x, and y are even.  
  
## Answer  
To prove this iff proposition, we have to consider it bidirectionally.  
*solution*  
**Proposition 1:** If w, x, and y are even then z is even.
**Lemma:** Even number's square is a even number, odd number's suqare is a odd number.  
The lemma tells that $w^{2},x^{2},y^{2}$ are even numbers, and since $w^{2}+x^{2}+y^{2}=z^{2}$, $z^{2}$ is a even number. By the lemma again, z
cannot be a odd number cause its square is even.  
**Proposition 2:** If z is even then w, x, and y are even. 
We can show by the lemma above that if one or three of w, x, and y are odd then $w^{2}+x^{2}+y^{2}$ is odd, but $z^{2}$ is even. By contradiction, we know that's not possible.  
So, here only left discussion for two of w, x, and y are odd. For convenience, we denote $z = 2l, w = 2i+1, x = 2j+1, y = 2k$. Substitute into the supposed equation, we get $$4i^{2}+4i+1+4j^{2}+4j+1+4k^{2}=4l^{2}$$  
Divided both sides by 2, we get$$2i^{2}+2i+2j^{2}+2j+2k^{2}+1 = 2l^{2}$$  
See that the left side is odd but right side is even.  
By contradiction, it's also not possible. So, w, x, and y must all be even.$$Q.E.D.$$
