# Problem 1
## Question
Define a *3-chain* to be a (not necessarily contiguous) subsequence
of three integers, which is either monotonically increasing or monotonically decreasing. We
will show here that any sequence of five distinct integers will contain a *3-chain*. Write
the sequence as a1, a2, a3, a4, a5. Note that a monotonically increasing sequences is one in
which each term is greater than or equal to the previous term. Similarly, a monotonically
decreasing sequence is one in which each term is less than or equal to the previous term.
Lastly, a subsequence is a sequence derived from the original sequence by deleting some
elements without changing the location of the remaining elements.  
  
**(a)** Assume that $a_1<a_2$. Show that if there is no *3-chain* in our sequence, then $a_3$ must be less than a1. (Hint: consider a4!)  
**(b)** Using the previous part, show that if $a_1<a_2$ and there is no *3-chain* in our
sequence, then $a_3 < a_4 < a_2$.  
**(c)** Assuming that $a_1<a_2$ and $a_3 < a_4 < a_2$， show that any value of $a_5$ must result in a *3-chain*.  
**(d)** Using the previous parts, prove by contradiction that any sequence of five distinct integers must contain a *3-chain*.  
  
## Answer
**(a)** *solution*: Since $a_1<a_2$, if no *3-chain* occurs, the following inequation must holds $a_3,a_4,a_5<a_2$, or $a_1,a_2,a_{3/4/5}$ is a *3-chain*. Similarly, $a_4>a_3$ must holds, or $a_2,a_3,a_4$ forms a *3-chain*. If $a_3>a_1$, then $a_1,a_3,a_4$ forms a *3-chain*. So, $$a_3<a_1 $$$$ Q.E.D.$$   
**(b)** *solution*: As we have already proved in **(a)**, $a_3,a_4,a_5<a_2$ and $a_4>a_3$ holds. We can combine those two inequations and get $a_3 < a_4 < a_2$. $$Q.E.D.$$  
**(c)** Now we should talk about $a_5$:  
*solution*: If $a_5<a_4$, then $a_2,a_4,a_5$ forms a *3-chain*, so $a_5>a_4$must holds. Then $a_3 < a_4 < a_5$ holds and forms a *3-chain*. So, whatever $a_5$ is, a *3-chain* forms.$$Q.E.D.$$  
**(d)** *solution*: Similarly, we can prove that if $a_1>a_2$ also always form a *3-chain* by substituting $<$ with $>$. Thereby, since $a_1\not= a_2$, one of the two conditions ($a_1<a_2 or a_1>a_2$) must holds and a *3-chain* is formed.$$Q.E.D.$$

# Problem 2
## Question
Prove by either the Well Ordering Principle or induction that for all nonnegative integers n:
$$\sum_{i=0}^{n}i^{3} = (\frac{n(n+1)}{2})^{2}$$  
  
## Answer
*solution*: **Base Case**: $n = 0$ $$left = \sum_{i=0}^{0}i^{3} = 0^{3} = 0$$ $$right = (\frac{0*(0+1)}{2})^{2} = 0$$ $$left = right$$  
**Induction Step**: Assume $P(n)$ holds.  
We need to prove: $$P(n+1): \sum_{i=0}^{n+1}i^{3} = (\frac{(n+1)(n+2)}{2})^{2}$$  
$$\sum_{i=0}^{n+1}i^{3} = \sum_{i=0}^{n}i^{3}+(n+1)^{3} = (\frac{n(n+1)}{2})^{2}+(n+1)^{3}$$  
$\Rightarrow$ $$(\frac{n(n+1)}{2})^{2}+(n+1)^{3} = \frac{n^{2}(n+1)^{2}}{4}+(n+1)^{3}$$
$$=(n+1)^2 \times\frac{n^2+4(n+1)}{4} = (\frac{(n+1)(n+2)}{2})^{2}$$
$$P(n)\Rightarrow P(n+1)$$
$$Q.E.D.$$

# Problem 4
## Question
Find the flaw in the following *bogus* proof that an = 1 for all
nonnegative integers n, whenever a is a nonzero real number.  
*Proof*. The *bogus* proof is by induction on n, with hypothesis
$$P(n) ::= \forall k \leq n. a^k = 1,$$
where $k$ is a nonnegative integer valued variable.  
**Base Case:** $P(0)$ is equivalent to $a^0 = 1$, which is true by definition of $a^0$. (By convention,
this holds even if a = 0.)  
**Inductive Step:** By induction hypothesis, $a^k = 1$ for all $k \in \Bbb{N}$ such that $k\leq n$. But then
$$a^{n+1} = \frac{a^n\cdot a^n}{a^{n-1}} = \frac{1\cdot 1}{1} = 1$$
which implies that $P(n + 1)$ holds. It follows by induction that $P(n)$ holds for all $n \in \Bbb{N}$,
and in particular, $a^n = 1$ holds for all $n \in N$.  
  
## Answer
*solution* The induction step involves the lastest two predicate to be true. $$a^{n+1} = \frac{a^n\cdot a^n}{a^{n-1}}$$
This means the base case should have two cases which is $n=0$ and $n=1$ both hold, while $P(1)$ doesn't hold.  
Thereby, the induction is bogus.

# Problem 5
## Question
Let the sequence $G_0,G_1,G_2,...$ be defined recursively as follows: $G_0 = 0,G_1 = 1,and G_n = 5G_{n-1}-6G_{n-2}$ for every $n\in\Bbb N,n\geq 2$  
Prove that for all $n\in\Bbb N, G_n = 3^n-2^n.$  
  
## Answer
We can easily prove this by strong induction or get the solution by solving the difference equation.  
*solution* **Base Case**: $$n=2\Rightarrow G_2 = 5\times G_1 - 6\times G_0 = 5$$
$$= 3^2-2^2  = 5$$ $$Q.E.D.$$  
**Inductive Step**: Suppose that $P(n)\ and\ P(n-1)$ holds, then we can use the recursive definition to get $G_{n+1}$:  
$$G_{n+1}= 5G_{n}-6G_{n-1}$$ $$ = 5\cdot 3^n-5\cdot 2^n-(6\cdot 3^{n-1}-6\cdot 2^{n-1})$$  
$$ = 3\cdot 3^n+2\cdot 2^n = 3^{n+1}+2^{n+1}$$ $$Q.E.D.$$

# Problem 7
## Question
There are two types of creature on planet Char, Z-lings and B-lings. Furthermore, every creature belongs to a particular generation. The creatures in each generation reproduce according to certain rules and then die off. The subsequent generation
consists entirely of their offspring.  
The creatures of Char pair with a mate in order to reproduce. First, as many Z-B pairs as
possible are formed. The remaining creatures form Z-Z pairs or B-B pairs, depending on
whether there is an excess of Z-lings or of B-lings. If there are an odd number of creatures,
then one in the majority species dies without reproducing. The number and type of offspring
is determined by the types of the parents  
$\cdot$ If both parents are Z-lings, then they have three Z-ling offspring.  
$\cdot$ If both parents are B-lings, then they have two B-ling offspring and one Z-ling offspring.  
$\cdot$ If there is one parent of each type, then they have one offspring of each type.  
There are 200 Z-lings and 800 B-lings in the first generation. Use induction to prove that
the number of Z-lings will always be at most twice the number of B-lings.  
*Hint: You may want to use a stronger hypothesis for the induction.*  
  
## Answer
**Denotation** If we consider the number of these two creatures as a state machine. **<Z,B>** is the state, where the first item **Z** denotes the number of Z-lings, so does the second item. For example the initial state is <200,800>.  
**Theorem**: The number of Z-lings is at most the number of B-lings.  
We use induction on the number of steps to prove ths theorem.  
*solution*: There are 3 cases for a single step of the state.  
(i) $Z=B$: In this case, every pair is Z-B pair. According to the third rule, every Z-B pair reproduces a Z-B pair. So the state is still **<Z,B>** .      
(ii) $Z<B$: In this case, there will be B-B pairs. According to the second rule, the new state is <$Z+\lfloor\frac{B-Z}{2}\rfloor,Z+2\times\lfloor\frac{B-Z}{2}\rfloor$>  
(iii) $Z>B$: In this case, there will be Z-Z pairs. According to the first rule, the new state is <$B+3\times\lfloor\frac{Z-B}{2}\rfloor,B$>  
**Base Case**: $n=0$ is the initial state, where $Z = 200<B = 800$  
**Inductive Step**: At the n step, $Z\leq B$, we need to prove that at the next state, the theorem still holds.  
$$Z_{n+1} = Z+\lfloor\frac{B-Z}{2}\rfloor\leq Z+\frac{B-Z}{2} = \frac{B+Z}{2}\leq 2\times B $$ $$Q.E.D.$$  
Specially, if you discuss the parity cases of $\lfloor\frac{B-Z}{2}\rfloor$, we can even prove our theorem.  
(i) If $B-Z$ is even then: $$Z+\lfloor\frac{B-Z}{2}\rfloor=Z+\frac{B-Z}{2} = \frac{B+Z}{2}$$
$$\because Z\leq B$$ $$\therefore \frac{B+Z}{2}\leq\frac{B+B}{2}=B$$  
(ii) If $B-Z$ is odd then: $$Z+\lfloor\frac{B-Z}{2}\rfloor=Z+\frac{B-Z-1}{2} = \frac{B+Z-1}{2}$$  
$$\because Z\leq B-1$$ $$\therefore \frac{B+Z-1}{2}\leq\frac{B+B-2}{2}=B-1$$  
$$\because Z+2\times\lfloor\frac{B-Z}{2}\rfloor \geq B-1$$ $$\therefore B_{n+1}\geq Z_{n+1}$$ $$Q.E.D.$$
